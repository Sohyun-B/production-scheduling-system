```python
import pandas as pd
from datetime import datetime
import ast

# 필요한 모듈 및 파이프라인 함수 import
from preprocessing import preprocessing
from yield_management import yield_prediction  # 변경된 함수명 반영
from dag_management import run_dag_pipeline, make_process_table
from scheduler import run_schedule, run_reschedule, DrawChart
from results import create_results

base_date = datetime(2025, 5, 15)  # 기준 날짜, 실제 환경에 맞게 변경
window_days = 5


# 0. 입력 데이터 준비 (예: 파일 로딩, DB 조회 등)
# 실제 경로/데이터 소스에 맞게 수정하세요
linespeed = pd.read_excel("품목별 분리 라인스피드 및 공정 순서.xlsx", sheet_name="품목별 라인스피드")
operation_seperated_sequence = pd.read_excel("품목별 분리 라인스피드 및 공정 순서.xlsx", sheet_name="공정순서")
# 기계기준정보
machine_master_info = pd.read_excel("품목별 분리 라인스피드 및 공정 순서.xlsx", sheet_name="기계기준정보")

# 수율 관련 데이터
yield_data = pd.read_excel("품목별 분리 라인스피드 및 공정 순서.xlsx", sheet_name="수율데이터")
operation_sequence = pd.read_excel("품목별 분리 라인스피드 및 공정 순서.xlsx", sheet_name="GITEM별 공정")

operation_types = pd.read_excel("공정 재분류 내역 및 교체 시간 정리(250820).xlsx", sheet_name="공정군")
operation_delay_df = pd.read_excel("공정 재분류 내역 및 교체 시간 정리(250820).xlsx", sheet_name="공정교체시간")
width_change_df = pd.read_excel("공정 재분류 내역 및 교체 시간 정리(250820).xlsx", sheet_name="폭변경")

machine_rest = pd.read_excel("불가능한 공정 입력값.xlsx", sheet_name = "기계")
machine_allocate = pd.read_excel("불가능한 공정 입력값.xlsx", sheet_name = "공정강제할당")
machine_limit = pd.read_excel("불가능한 공정 입력값.xlsx", sheet_name = "공정강제회피")



# 주문정보 간략 변경
order = pd.read_excel('25년 5월 PO 내역(송부건).xlsx', sheet_name = '25년 5월 PO확정 내역')
# order = order[order['납기일'].dt.month == 5] # 5월 주문만 필터링
order[['두께', '너비', '길이']] = order['SPEC'].str.split("*", expand=True).astype(float)
order.loc[order['너비'] == 914.4, '너비'] = 914
order['원단길이']= order['길이'] * order['의뢰량']
order = order[['P/O NO', 'GITEM', 'GITEM명', '너비', '길이', '의뢰량', '원단길이', '납기일']]
```

```python
# 1. 데이터 전처리
sequence_seperated_order, linespeed = preprocessing(order, operation_seperated_sequence, operation_types, machine_limit, machine_allocate, linespeed)
# print("+++++++++++sequence_seperated_order++++++++++")
# display(sequence_seperated_order)

# 2. 수율 예측 + 수율 매핑에 따른 생산길이 조정 (변경된 함수 사용)
yield_predictor, sequence_yield_df, sequence_seperated_order = yield_prediction(
    yield_data,
    operation_sequence,
    sequence_seperated_order
)

# 3. DAG 생성 및 그래프 매니저, 기계 딕셔너리 생성
merged_df = make_process_table(sequence_seperated_order)

hierarchy = sorted(
    [col for col in merged_df.columns if col.endswith('ID')],
    key=lambda x: int(x.split('공정')[0])  # 숫자 기준 정렬 (수정 가능)
)


dag_df, opnode_dict, manager, machine_dict = run_dag_pipeline(
    merged_df,
    hierarchy,
    sequence_seperated_order,
    linespeed,
    machine_columns=machine_master_info['기계코드'].values.tolist()
)



# 4. 스케줄링 실행
output_final_result, manager, scheduler = run_schedule(
    dag_df,
    sequence_seperated_order,
    machine_dict,
    manager,
    window_days,
    opnode_dict,
    machine_rest, 
    base_date,
    operation_delay_df, 
    width_change_df
)

# depth -1인 가짜 공정 삭제
output_final_result = output_final_result[ ~(output_final_result['depth']== -1)]

machine_schedule_df=scheduler.create_machine_schedule_dataframe()

# 할당 작업이 -1 공정인 경우, '[-1'로 시작하는 경우 삭제
machine_schedule_df = machine_schedule_df[ ~machine_schedule_df['할당 작업'].astype(str).str.startswith('[-1', na=False)]

# 5. 결과 처리
results = create_results(
    output_final_result=output_final_result,
    merged_df=merged_df,
    original_order=order,
    sequence_seperated_order=sequence_seperated_order,
    machine_mapping = machine_master_info.set_index('기계인덱스')['기계코드'].to_dict(),
    machine_schedule_df=machine_schedule_df,
    base_date=base_date,
    scheduler=scheduler
)

gantt = DrawChart(scheduler.Machines)
display(gantt.plot())


# 출력 예시
print(f"지각 총 일 수: {results['late_days_sum']}")
print(f"총 소요시간: {results['new_output_final_result']['종료시각'].max() / 48}일")
print(f"총 makespan: {results['new_output_final_result']['종료시각'].max()}")

def result_to_gantt_data(original_output, output_final_result):
    name_map = {"ID" : "id",
     "작업 시작 시간" : "start",
     "작업 종료 시간" : "end",
     "공정순서" : "depth",
     "GITEM" : "item",
     "배합코드" : "mixture",
     "호기": "machine"
    }
    original_output = original_output.rename(columns = name_map)
    original_output['content'] = (
        original_output['id']
            .str.split('_')           # split into list by underscore
            .str[:2]                  # take first two parts
            .str.join('_')            # join back with underscore
    )
    original_output['operation_name'] = (
        original_output['id']
            .str.split('_')           # split into list by underscore
            .str[1]                  # take first two parts
    )
    machine_mapping = machine_master_info.set_index('기계코드')['기계이름'].to_dict()
    # machine_mapping={'C2010': 1, 'C2250': 25, 'C2260': 26, 'C2270': 27, 'O2310':31, 'O2340':34, 'O2360':36, 'O2510':51}
    original_output['machine'] = original_output['machine'].map(machine_mapping)
    
    original_output = pd.merge(original_output, output_final_result, on ='id', how = 'left')
    original_output = original_output[['id', 'content', 'start', 'end', 'machine', 'depth', 'operation_name', 'item', 'mixture', 'children']]
    
    parent_map = {}
    def process_row(row):
        val = row['children']
        # 리스트화
        children = ast.literal_eval(val) if isinstance(val, str) else val
        for child in children:
            parent_map[child] = row['id']
        return children

    # children을 리스트로 변환
    original_output['children'] = original_output.apply(lambda row: process_row(row), axis=1)
    # 리스트를 쉼표로 연결된 문자열로 변환
    original_output['children'] = original_output['children'].apply(lambda x: ','.join(x) if x else '')

    # parent 컬럼 추가
    original_output['parent'] = original_output['id'].map(parent_map).fillna('')

    display(original_output)
    original_output.to_excel("gantt_input.xlsx")
    return original_output    
result_to_gantt_data(results['machine_info'], output_final_result[['id', 'children']])

# 엑셀 파일에 저장
with pd.ExcelWriter("0821결과.xlsx", engine="openpyxl") as writer:
    results['new_output_final_result'].to_excel(writer, sheet_name="주문_생산_요약본", index=False)
    results['machine_info'].to_excel(writer, sheet_name="호기_정보", index=False)
    results['merged_result'].to_excel(writer, sheet_name="주문_생산_정보", index=False)

############### reschedule ########

# # 1, 25, 26 등으로 되어있는 machine을 1호기, 25호기 등으로 변경 필요
# gantt_result = pd.read_excel("updated_timeline_data (2).xlsx", sheet_name = "Timeline Data")

# def gantt_data_to_result(gantt_result):
#     # 1, 25, 26 등으로 되어있는 machine을 1호기, 25호기 등으로 변경 필요
#     # machine_mapping = machine_master_info.set_index('기계이름')['기계인덱스'].to_dict()
#     machine_mapping={1: 0, 25: 1, 26: 2, 27: 3, 31:4, 34:5, 36:6, 51:7}
#     gantt_result['machine'] = gantt_result['machine'].map(machine_mapping)

#     result = (
#         gantt_result
#         .sort_values(['machine', 'start'])       # 기계별로 시작시간 정렬
#         .groupby('machine')['id']                # 기계별로 id 묶기
#         .apply(list)                              # list로 변환
#         .reset_index(name='id_list')              # 보기 좋게 컬럼명 변경
#     )

#     machine_queues = dict(zip(result['machine'], result['id_list']))
#     return machine_queues
# machine_queues = gantt_data_to_result(gantt_result)


# #     new_output_final_result = run_reschedule(machine_queues, manager)
# new_output_final_result = run_reschedule(scheduler, machine_queues, manager, machine_rest, base_date)

# print("리스케줄 이후")
# display(new_output_final_result)
# print(new_output_final_result['parent_node_count'].value_counts())
# print(new_output_final_result['processing_time'].isna().sum())
    
# gantt = DrawChart(scheduler.Machines)
# display(gantt.plot())
```
